#### Import needed modules

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import zipfile
import os
from math import log
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Unzip and load datasets (being careful of types)

```python
with zipfile.ZipFile('kc_house_data.csv.zip', "r") as z:
    z.extractall(os.getcwd())
with zipfile.ZipFile('kc_house_test_data.csv.zip', "r") as z:
    z.extractall(os.getcwd())
with zipfile.ZipFile('kc_house_train_data.csv.zip', "r") as z:
    z.extractall(os.getcwd())
    
# Dictionary with the correct dtypes for the DataFrame columns
dtype_dict = {'bathrooms':float, 'waterfront':int, 'sqft_above':int, 
              'sqft_living15':float, 'grade':int, 'yr_renovated':int, 
              'price':float, 'bedrooms':float, 'zipcode':str, 
              'long':float, 'sqft_lot15':float, 'sqft_living':float, 
              'floors':str, 'condition':int, 'lat':float, 'date':str, 
              'sqft_basement':int, 'yr_built':int, 'id':str, 'sqft_lot':int, 'view':int}
    
sales = pd.read_csv('kc_house_data.csv', dtype = dtype_dict)
train_data = pd.read_csv('kc_house_train_data.csv', dtype = dtype_dict)
test_data = pd.read_csv('kc_house_test_data.csv', dtype = dtype_dict)
```

#### Preview of train data

```python
train_data.head()
```

#### Define dependant features

```python
train_data['bedrooms_squared'] = train_data['bedrooms']*train_data['bedrooms']
train_data['bed_bath_rooms'] = train_data['bedrooms']*train_data['bathrooms']
train_data['log_sqft_living'] = np.log(train_data['sqft_living'])
train_data['lat_plus_long'] = train_data['lat']+train_data['long']

test_data['bedrooms_squared'] = test_data['bedrooms']*test_data['bedrooms']
test_data['bed_bath_rooms'] = test_data['bedrooms']*test_data['bathrooms']
test_data['log_sqft_living'] = np.log(test_data['sqft_living'])
test_data['lat_plus_long'] = test_data['lat']+test_data['long']
```

```python
print("Test data means\nbedrooms_squared: {0}\nbed_bath_rooms: {1}\nlog_sqft_living: {2}\nlat_plus_long: {3}\n".format(
    test_data['bedrooms_squared'].mean(),
    test_data['bed_bath_rooms'].mean(),
    test_data['log_sqft_living'].mean(),
    test_data['lat_plus_long'].mean()))
```

### Training a multiple regression model

Lets estimate the regression coefficients/weights for predicting ‘price’ for the following three models:(In all 3 models include an intercept -- most software does this by default).

**Model 1**: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’, and ‘long’

**Model 2**: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’,‘long’, and ‘bed_bath_rooms’

**Model 3**: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’,‘long’, ‘bed_bath_rooms’, ‘bedrooms_squared’, ‘log_sqft_living’, and ‘lat_plus_long’

```python
# How the relation between those variables and price look like separately
plt.figure(figsize=(18,18))
plt.subplot(3, 3, 1)
plt.plot(train_data['sqft_living'], train_data['price'],'g.')
plt.xlabel('Living Area (ft^2)', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)
plt.grid()

plt.subplot(3, 3, 2)
plt.plot(train_data['bedrooms'], train_data['price'],'g.')
plt.xlabel('# Bedrooms', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.subplot(3, 3, 3)
plt.plot(train_data['bathrooms'], train_data['price'],'g.')
plt.xlabel('# Bathrooms', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.subplot(3, 3, 4)
plt.plot(train_data['lat'], train_data['price'],'g.')
plt.xlabel('Latitude', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)
plt.grid()

plt.subplot(3, 3, 5)
plt.plot(train_data['long'], train_data['price'],'g.')
plt.xlabel('Longitud', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.subplot(3, 3, 6)
plt.plot(train_data['bed_bath_rooms'], train_data['price'],'g.')
plt.xlabel('Bedrooms x Bathrooms', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.subplot(3, 3, 7)
plt.plot(train_data['bedrooms_squared'], train_data['price'],'g.')
plt.xlabel('# Bedrooms^2', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)
plt.grid()

plt.subplot(3, 3, 8)
plt.plot(train_data['log_sqft_living'], train_data['price'],'g.')
plt.xlabel('log(sq.feet)', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.subplot(3, 3, 9)
plt.plot(train_data['lat_plus_long'], train_data['price'],'g.')
plt.xlabel('Lat + Long', fontsize=10)
plt.ylabel('House Price ($)', fontsize=10)

plt.show()
```

Create LinearRegression object and fit the model

```python
# Model 1
example_features_model1 = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']
X_multilin_model1 = train_data[example_features_model1]
Y_multilin = train_data['price']
multilin_obj_model1 = LinearRegression(copy_X=True, normalize=False, n_jobs=1, fit_intercept=True)
multilin_obj_model1.fit(X_multilin_model1, Y_multilin)

# Model 2
example_features_model2 = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long',
                           'bed_bath_rooms']
X_multilin_model2 = train_data[example_features_model2]
multilin_obj_model2 = LinearRegression(copy_X=True, normalize=False, n_jobs=1, fit_intercept=True)
multilin_obj_model2 = multilin_obj_model2.fit(X_multilin_model2, Y_multilin)

# Model 3
example_features_model3 = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms',
                           'bedrooms_squared', 'log_sqft_living', 'lat_plus_long']
X_multilin_model3 = train_data[example_features_model3]
multilin_obj_model3 = LinearRegression(copy_X=True, normalize=False, n_jobs=1, fit_intercept=True)
multilin_obj_model3 = multilin_obj_model3.fit(X_multilin_model3, Y_multilin)
```

Now, let's gather the computed coefficients and intercept

```python
print("Model 1\nIntercept: {0}\nCoefficients: {1}".format(multilin_obj_model1.intercept_, multilin_obj_model1.coef_))
print("Model 2\nIntercept: {0}\nCoefficients: {1}".format(multilin_obj_model2.intercept_, multilin_obj_model2.coef_))
print("Model 3\nIntercept: {0}\nCoefficients: {1}".format(multilin_obj_model3.intercept_, multilin_obj_model3.coef_))
```

*What is the sign (positive or negative) for the coefficient/weight for ‘bathrooms’ in Model 1?*

Positive

*What is the sign (positive or negative) for the coefficient/weight for ‘bathrooms’ in Model 2?*

Negative

*Is the sign for the coefficient the same in both models? Think about why this might be the case.*

In model 2, the new 'bed_bath_rooms' feature causes the house price to be over-estimated. Thus, the 'bathrooms' turns to negative to better agree with the observed values for the house prices.

#### Model comparing

```python
def get_residual_sum_of_squares(model, data, outcome):
    pred = model.predict(data)
    res = pred - outcome
    RSS = sum(res**2)
    return(RSS)   
```

Calculate the RSS for the 3 models over the **test data**

```python
rss_multilin_obj_model1 = get_residual_sum_of_squares(multilin_obj_model1, test_data[example_features_model1], test_data['price'])
rss_multilin_obj_model2 = get_residual_sum_of_squares(multilin_obj_model2, test_data[example_features_model2], test_data['price'])
rss_multilin_obj_model3 = get_residual_sum_of_squares(multilin_obj_model3, test_data[example_features_model3], test_data['price'])
print("RSS Model 1: {0}\nRSS Model 2: {1}\nRSS Model 3: {2}\n".format(rss_multilin_obj_model1,rss_multilin_obj_model2,rss_multilin_obj_model3))
```

Calculate the RSS for the 3 models over the **train data**

```python
rss_multilin_obj_model1 = get_residual_sum_of_squares(multilin_obj_model1, train_data[example_features_model1], train_data['price'])
rss_multilin_obj_model2 = get_residual_sum_of_squares(multilin_obj_model2, train_data[example_features_model2], train_data['price'])
rss_multilin_obj_model3 = get_residual_sum_of_squares(multilin_obj_model3, train_data[example_features_model3], train_data['price'])
print("RSS Model 1: {0}\nRSS Model 2: {1}\nRSS Model 3: {2}\n".format(rss_multilin_obj_model1,rss_multilin_obj_model2,rss_multilin_obj_model3))
```

**Which model (1, 2 or 3) had the lowest RSS on TRAINING data?**
> Model 3 has the lowest RSS on the Training Data. This is expected since Model 3 has the most features, but perhaps it will be overfitted.

**Which model (1, 2, or 3) had the lowest RSS on TESTING data?**
> Model 2 has the lowest RSS. Probably because there is a reinforcement in the bedrooms x bathrooms feature, which have a low error separately when estimating price.
